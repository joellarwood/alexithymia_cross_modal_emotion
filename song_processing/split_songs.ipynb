Select songs that meet criterea

```python
anger = ratings[ratings['TARGET'].str.contains("ANGER")]

# I need the anger rating to be higher than the ratings for all othe emotions 

anger = anger.query('Anger > Fear')

anger = anger.assign(dist = anger["Anger"] - anger["Fear"]) 

anger = anger[anger.dist > .5]
```

### Fear

```python
fear = ratings[ratings['TARGET'].str.contains("FEAR")]

# I need the anger rating to be higher than the ratings for all othe emotions 

fear = fear.query('Fear > Anger')

fear = fear.assign(dist = fear["Fear"] - fear["Anger"]) 

fear = fear[fear.dist > .5]

fear_tracks = fear["Number"].tolist()

fear
```

```python
happy = ratings[ratings["TARGET"].str.contains("HAPPY")]

happy = happy.query('Happy > Sad')

happy = happy.query("Happy > Tender")

happy = happy.assign(dist = happy["Happy"] - happy["Sad"])

happy = happy[happy.dist > .5]

happy_track = happy["Number"].tolist()

happy
```

### Sad
